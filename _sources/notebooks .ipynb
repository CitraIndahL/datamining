**2. Membuat Statistik Deskriptif dari data (setiap atribut yang bertipe numerik)**

```python
med = {'Confirmed': df['Confirmed'].median(),'Recovered':df['Recovered'].median(),
    'Deaths':df['Deaths'].median()
    }

df = pd.DataFrame(med, index=['median'])
df
```

```python
rata2 = {'Confirmed': df['Confirmed'].mean(),'Recovered':df['Recovered'].mean(),
    'Deaths':df['Deaths'].mean()
    }

df = pd.DataFrame(rata2, index=['mean'])
df
```

```python
modus= {'Confirmed': df['Confirmed'].mode(),'Recovered':df['Recovered'].mode(),
    'Deaths':df['Deaths'].mode()
    }
df = pd.DataFrame(modus, index=['mode'])
df
```

**3. SKEWED / KEMIRINGAN**

```python
sns.distplot(df['Confirmed'],kde=True,
             bins= int(150/5),color='red',
             hist_kws={'edgecolor':'black'},
             kde_kws={'linewidth':7}).set(title='Positive Skewed')
```

```python
sns.distplot(df['Recovered'],kde=True,
             bins= int(150/5),color='darkblue',
             hist_kws={'edgecolor':'black'},
             kde_kws={'linewidth':4}).set(title='Positive Skewed')
```

```python
sns.distplot(df['Deaths'],kde=True,
             bins= int(150/5),color='darkblue',
             hist_kws={'edgecolor':'black'},
             kde_kws={'linewidth':4}).set(title='Positive Skewed')
```
