 **Metode KNN**

```python
import pandas as pd
import numpy as np
```

```python
dataset_url="https://raw.githubusercontent.com/CitraIndahL/datamining/main/dataR2.csv"
data=pd.read_csv(dataset_url, index_col=0)
```

```python
data
```

```python
#Check the number of rows and columns in the data
data.shape
```

```python
#create a dataframe by excluding the target column
X = data.drop(columns=['Classification'])
#check by displaying data
X.head()
```

```python
#Target values
y = data['Classification'].values
#showing target values
y[0:5]
```

```python
from sklearn.model_selection import train_test_split
#separate dataset to train and test data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1, stratify=y)
```

```python
from sklearn.neighbors import KNeighborsClassifier
# Make KNN classifier
knn = KNeighborsClassifier(n_neighbors = 3)
# Customize class with data
knn.fit(X_train,y_train)
```

```python
#Show 5 prediction data
knn.predict(X_test)[0:5]
```

```python
#Check the accuracy of the test data
knn.score(X_test, y_test)
```

```python
from sklearn.model_selection import cross_val_score
import numpy as np
#Make KNN model baru
knn_cv = KNeighborsClassifier(n_neighbors=3)
#Check the accuracy of the test data, try the model with cv = 5
cv_scores = cross_val_score(knn_cv, X, y, cv=5)
#print cv score
print(cv_scores)
```

```python
print("cv_scores mean:{}".format(np.mean(cv_scores)))
```

```python
from sklearn.model_selection import GridSearchCV
#make a new KNN Model
knn2 = KNeighborsClassifier()
#create a dictionary of all values â€‹â€‹to be tested on n_neighbors
param_grid = {"n_neighbors": np.arange(1, 25)}
#use gridsearch to test values â€‹â€‹on n_neighbors
knn_gscv = GridSearchCV(knn2, param_grid, cv=5)
#adjust the model on the data
knn_gscv.fit(X, y)
```

```python
#check the performance of n_neighbors value
knn_gscv.best_params_
```

```python
#check the mean score of the top performing value on n_neighbors
knn_gscv.best_score_
```

```python
# calculate the error at K values â€‹â€‹between 2 and 10
error = []
import matplotlib.pyplot as plt
for i in range(1, 10):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train, y_train)
    pred_i = knn.predict(X_test)
    error.append(np.mean(pred_i != y_test))
plt.figure(figsize=(12, 6))
plt.plot(range(1, 10), error, color='red', linestyle='dashed', marker='o',
         markerfacecolor='blue', markersize=10)
plt.title('Error Rate K Value')
plt.xlabel('K Value')
plt.ylabel('Mean Error')
```

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
from sklearn import tree
import seaborn as sns
%matplotlib inline
```

```python
import warnings

warnings.filterwarnings('ignore')
```

```python
dataset_url="https://raw.githubusercontent.com/CitraIndahL/datamining/main/dataR2.csv"
df = pd.read_csv(dataset_url, index_col=0)
```

```python
# preview from dataset

df.head()
```

```python
df.info()
```

```python
col_names = ['BMI', 'Glucose', 'Insulin', 'HOMA', 'Leptin', 'Adiponectin', 'Resistin', 'MCP.1', 'Classification']


for col in col_names:
    
    print(df[col].value_counts())   
```

```python
df['Classification'].value_counts()
```

```python
# check if there are missing values â€‹â€‹in the variables

df.isnull().sum()
```

```python
#declare feature vector and target variable

X = df.drop(['Classification'], axis=1)
y = df['Classification']
```

```python
# split X and y into training and testing sets

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 42)
```

```python
# check the shape of X_train and X_test

X_train.shape, X_test.shape
```

```python
#perform a data type check on X_train

X_train.dtypes
```

```python
# import DecisionTreeClassifier

from sklearn.tree import DecisionTreeClassifier
```

```python
# instantiate the DecisionTreeClassifier model with criterion gini index

clf_gini = DecisionTreeClassifier(criterion='gini', max_depth=3, random_state=0)


# fit the model
clf_gini.fit(X_train, y_train)
```

```python
# print the scores on training and test set

print('Training set score: {:.4f}'.format(clf_gini.score(X_train, y_train)))

print('Test set score: {:.4f}'.format(clf_gini.score(X_test, y_test)))
```

```python
y = df["Classification"]
X = df.drop(columns=["Classification"])
clf = tree.DecisionTreeClassifier(criterion="gini")
clf = clf.fit(X, y)
```

```python
#plt the figure, setting a black background
plt.figure(figsize=(10,10))
#create the tree plot
a = tree.plot_tree(clf,
                   rounded = True,
                   filled = True,
                   fontsize=8)
#show the plot
plt.show()
```
